```python
#load the dataset
filename = 'data/NBA.csv' 
NBA = csv.reader(open(filename, newline=''), delimiter=',')

header = next(NBA) 
print(header)

dataset = list(NBA)
for i in range(len(dataset)):
    dataset[i] = [int(x) for x in dataset[i]]
    
dataset = np.asarray(dataset)

X = dataset[:,1:4] #columns 1,2,3 contain the features
Y = dataset[:,0]  # column 0: labels

Y = Y*2-1  

m = dataset.shape[0]
permutation = np.random.permutation(m) # random permurtation

X = X[permutation]
Y = Y[permutation]
print(X)
print(Y)
print(permutation)
```

```python
#the number of samples in the training set
training_m, test_m = 195, 65
m_training = training_m 

#the number of samples in the test set
m_test = test_m 

X_training = X[:195]   #instances for the training set
Y_training = Y[:195]   #labels for the training set


X_test = X[195:]   #instances for test set
Y_test = Y[195:]   #labels for the test set

print(X_training)
print(Y_training)
print(m_test)

print("Shape of training set: " + str(X_training.shape))
print("Shape of test set: " + str(X_test.shape))
```

```python
#add an 1 to each sample
arr_x1 = np.ones((195, 1))
Xh = np.hstack((X_training, arr_x1))
print(Xh)

arr_y1 = np.ones((65, ))
Yh = np.hstack((Y_training, arr_y1))
print(Yh)
print(len(Xh))
```

```python
def perceptron_update(current_w, x, y):
    new_w = current_w + y*x
    return new_w

def perceptron(X, Y, max_num_iterations):
    curr_w = np.zeros(X.shape[1])
    best_w = curr_w
    num_samples = X.shape[0]
    best_error = num_samples+1
    index_missclassified = -1
    num_missclassified = 0
    for i in range(num_samples):
        permutation = np.random.permutation(num_samples)
        x = X[permutation[i],:]
        y = Y[permutation[i]]
        prediction = np.sign(np.dot(curr_w,x))
        if (prediction != y):
            index_missclassified = permutation[i]
            num_missclassified += 1
    if num_missclassified < best_error:
        best_error = num_missclassified
        best_w = curr_w
    num_iter = 1
    while ((index_missclassified != -1) and (num_iter < max_num_iterations)):
        curr_w = perceptron_update(curr_w, X[index_missclassified,:],Y[index_missclassified])
        index_missclassified = -1
        num_missclassified = 0
        for i in range(num_samples):
            permutation = np.random.permutation(num_samples)
            x = X[permutation[i],:]
            y = Y[permutation[i]]
            prediction = np.sign(np.dot(curr_w,x))
            if (prediction != y):
                index_missclassified = permutation[i]
                num_missclassified += 1
        if num_missclassified < best_error:
            best_error = num_missclassified
            best_w = curr_w
        num_iter += 1
    best_error = float(best_error)/float(num_samples)
    return best_w, best_error
```

```python
#the perceptron for 100 iterations
w_found, error = perceptron(X_training,Y_training, 100)
print("Training Error of perpceptron (100 iterations): " + str(error))

#predictions on test dataset
num_errors = 0.

for i in range(m_test):
    x = X_test[i,:]
    y = Y_test[i]
    predicted = np.sign(np.dot(w_found,x))
    if (predicted != y):
        num_errors += 1


true_loss_estimate = num_errors/m_test  # error rate on the test set
print("Test Error of perpceptron (100 iterations): " + str(true_loss_estimate))
```
